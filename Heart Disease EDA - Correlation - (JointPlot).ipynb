```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('heart.csv')
df.head()
```

## Correlation - Joint Plot

```python
sns.jointplot(data = df, x = 'Age', y = 'MaxHR', kind = 'hex')
```

```python
sns.jointplot(data = df, x = 'Age', y = 'MaxHR', kind = 'reg')
```

`Here, we can see that according to reg-line, MaxHR is decrieasing as Age is incrieases `

```python
sns.jointplot(data = df, x = 'HeartDisease', y = 'MaxHR', kind = 'reg')
```
