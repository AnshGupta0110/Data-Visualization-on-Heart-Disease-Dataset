```python
import pandas as pd 
import seaborn as sns


df = pd.read_csv('heart.csv')
df.head()
```

# Distribution for Numerical columns

## RestingBP distribution

```python
sns.displot(df['RestingBP'], kde = False, color = 'blue')
```

```python
sns.distplot(df['RestingBP'], kde = True, color = 'blue')
```

## Distribution of Cholesterol

```python
sns.distplot(df['Cholesterol'], kde = True, color = 'blue')
```

## Distribution of MaxHR

```python
sns.distplot(df['MaxHR'], kde = True, color = 'blue')
```

# Pie Chart for categorical 

## ratio between man and female

```python
df.groupby('Sex').size()
```

```python
df.groupby('Sex').size().plot(kind = 'pie' , autopct = '%.2f')
```
